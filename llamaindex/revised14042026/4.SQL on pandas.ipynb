```python
1.0 Call libraries
from llama_index.embeddings.ollama import OllamaEmbedding
from llama_index.llms.ollama import Ollama
from llama_index.core import SimpleDirectoryReader
from llama_index.experimental.query_engine import PandasQueryEngine
from llama_index.core import Settings
import pandas as pd
```

```python
# 1.1 Initialise llms
llm = Ollama(
             model=  "codellama:13b",    # "deepseek-r1:7b" works best for router agent, # "qwen3.5:4b", #      "llama3.2:latest",
             request_timeout=3600,  # Change it
             temperature = 0,       # To produce stable deterministic output
             #context_window=,    # Limit timeouts
             mirostat = 0
             )

embed_model = OllamaEmbedding(
                                model_name="nomic-embed-text",      # Using foundational model may be overkill
                                base_url="http://localhost:11434",
                              )
```

```python
# 1.2 Read csv file
df = pd.read_csv("/home/ashok/Documents/llamaindex/deeplearning_course/csv_data/customers-10000.csv")

# 1.3
df.columns
df.head()
df.shape
```

```python
# 2. Initialize the query engine
# Setting verbose=True lets you see the generated pandas code
query_engine = PandasQueryEngine(df=df, 
                                 llm = llm, 
                                 verbose=True,             # see the generated pandas code
                                 synthesize_response=True  # Synthesizes Response: Puts the output in natural language
                                )

```

```python
# 3. Query the data in natural language   Aaronstad
# response = query_engine.query("Give me the 'FirstName' of all those who are in City of 'East Leonard'")
response = query_engine.query("Give me the 'FirstName' of all those who are in City of 'Aaronville'")
print(response)
```
